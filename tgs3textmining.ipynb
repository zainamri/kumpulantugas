```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')

stop_words = set(stopwords.words('indonesian'))

factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Membuka data yang sudah ditarik dari YouTube
df = pd.read_csv('/content/komentar_deforestasipapua.csv')

print("Data Asli (Sebelum Dibersihkan):")
display(df[['Teks Komentar']].head())
```

```python
# Mengubah semua huruf menjadi kecil
df['text_lower'] = df['Teks Komentar'].astype(str).str.lower()

print("Hasil Case Folding:")
display(df[['Teks Komentar', 'text_lower']].head())
```

```python
# Fungsi untuk menghapus selain huruf a-z dan spasi
def clean_punctuation(text):
    return re.sub(r'[^a-z\s]', '', text)

df['text_clean'] = df['text_lower'].apply(clean_punctuation)

print("Hasil Cleaning Tanda Baca:")
display(df[['text_lower', 'text_clean']].head())
```

```python
# Kamus bahasa gaul
slang_dict = {
    "gk": "tidak",
    "ga": "tidak",
    "gak": "tidak",
    "bgt": "banget",
    "banget": "sangat",
    "yg": "yang",
    "klo": "kalau",
    "tp": "tapi",
    "thn": "tahun",
    "dgn": "dengan"
}

def normalize_slang(text):
    words = text.split()
    return " ".join([slang_dict.get(word, word) for word in words])

df['text_slang'] = df['text_clean'].apply(normalize_slang)

print("Hasil Perbaikan Bahasa Gaul:")
display(df[['text_clean', 'text_slang']].head())
```

```python
# Memotong kalimat menjadi kata per kata
df['tokens'] = df['text_slang'].apply(word_tokenize)

print("Hasil Tokenisasi:")
display(df[['text_slang', 'tokens']].head())
```

```python
# Membuang kata tidak penting
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

df['tokens_stop'] = df['tokens'].apply(remove_stopwords)

print("Hasil Stopword Removal:")
display(df[['tokens', 'tokens_stop']].head())
```

```python
# Mencari kata dasar
def stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

print("Mesin sedang mengupas kata... tunggu sebentar ya!")
df['tokens_stem'] = df['tokens_stop'].apply(stemming)

print("\n Hasil Stemming:")
display(df[['tokens_stop', 'tokens_stem']].head())
```

```python
# Menyatukan kata menjadi teks utuh kembali
df['final_text'] = df['tokens_stem'].apply(lambda x: ' '.join(x))

print("Hasil Akhir Teks Bersih:")
display(df[['Teks Komentar', 'final_text']].head())

# Menyimpan data untuk dilabeli secara manual
df_untuk_dilabeli = df[['Teks Komentar', 'final_text']]
df_untuk_dilabeli.to_csv('data_siap_label.csv', index=False)

print("\nBerhasil! File 'data_siap_label.csv' sudah tersimpan di folder sebelah kiri.")
```

```python
import pandas as pd

# Membuka data yang tadi
df = pd.read_csv('data_siap_label.csv')

# Membuat kamus sederhana untuk penilaian otomatis
pos_words = ['bagus', 'baik', 'cerdas', 'suka', 'mantap', 'menarik', 'hebat', 'keren', 'manfaat', 'setuju', 'luar biasa', 'recommended', 'penting', 'edukasi', 'cerah', 'maju', 'pintar', 'sukses', 'semangat', 'paham', 'nyambung', 'hormat', 'dukung', 'terbaik', 'solusi', 'bantu', 'berguna', 'wow']
neg_words = ['buruk', 'jelek', 'babi', 'kecewa', 'parah', 'lama', 'bodoh', 'tolol', 'goblok', 'bahaya', 'takut', 'ancam', 'kurang', 'salah', 'kacau', 'hancur', 'ngeri', 'pusing', 'susah', 'berat', 'bingung', 'tolak', 'anti', 'ancaman', 'rugi', 'gagal', 'sulit', 'malas']

# Fungsi penilai skor
def auto_label(text):
    if not isinstance(text, str):
        return 0
    words = text.split()

    # Hitung jumlah kata positif dan negatif
    pos_score = sum(1 for w in words if w in pos_words)
    neg_score = sum(1 for w in words if w in neg_words)

    # Beri label 1 (Positif) jika kata positif lebih banyak, sisanya 0 (Negatif)
    return 1 if pos_score >= neg_score else 0

# Terapkan fungsi pelabelan otomatis ke kolom 'final_text'
df['label'] = df['final_text'].apply(auto_label)

print("Proses pelabelan otomatis selesai!")
display(df[['final_text', 'label']].head(10))

# Menyimpan data yang sudah berlabel
df.to_csv('data_dilabeli.csv', index=False)
print("\nBagus! File 'data_dilabeli.csv' sudah berhasil dibuat.")
```
