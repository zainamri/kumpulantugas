```python
import pandas as pd
from googleapiclient.discovery import build
```

```python
api_key = "AIzaSyCCtyncn4F7iqE1SOPRK0cWh24b3IaUWuc"
video_id = "10DCZWqAsdI"

# mengkoneksikan ke YouTube API
youtube = build('youtube', 'v3', developerKey=api_key)
```

```python
# PROSES PENARIKAN DATA KOMENTAR (API)
komentar_list = []
next_page_token = None

print("Mulai menarik data komentar...")

while len(komentar_list) < 150:
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=100,
        pageToken=next_page_token,
        textFormat="plainText"
    )
    response = request.execute()
    for item in response['items']:
        komentar = item['snippet']['topLevelComment']['snippet']['textDisplay']
        komentar_list.append(komentar)
    next_page_token = response.get('nextPageToken')

    if not next_page_token:
        break
```

```python
df_komentar = pd.DataFrame({'Teks Komentar': komentar_list})

print(f"\nSelesai! Berhasil menarik {len(df_komentar)} data teks komentar.")

df_komentar.to_csv('komentar_deforestasipapua.csv', index=False, encoding='utf-8')
print("Data telah disimpan dengan nama 'komentar_DEFORESTASI PAPUA DEMI FOOD ESTATE.csv'")

df_komentar.head()
```
